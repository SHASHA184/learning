# Facade Pattern

## Intent
Provide a unified, simplified interface to a set of interfaces in a subsystem. Facade defines a higher-level interface that makes the subsystem easier to use.

## Problem
You have a complex subsystem with many interconnected classes:
- Too many classes to interact with
- Complex initialization sequences
- Difficult to understand dependencies
- Clients shouldn't know implementation details

**Real-world analogy**: Restaurant waiter is a facade - you don't interact with chefs, dishwashers, and managers directly.

## When to Use
âœ… **Use when:**
- Need simple interface to complex subsystem
- Many dependencies between clients and implementation classes
- Want to layer subsystems
- Want to hide complexity from clients

âŒ **Avoid when:**
- Subsystem is already simple
- Clients need fine-grained control
- Facade becomes too large (God object)

## Example 1: Home Theater (Without Facade)

**Problem**: Too many steps to watch a movie

````python
# WITHOUT Facade - Complex subsystem

class DVDPlayer:
    def on(self):
        print("DVD Player: Turning on")
    
    def play(self, movie):
        print(f"DVD Player: Playing '{movie}'")
    
    def stop(self):
        print("DVD Player: Stopping")
    
    def off(self):
        print("DVD Player: Turning off")

class Projector:
    def on(self):
        print("Projector: Turning on")
    
    def wide_screen_mode(self):
        print("Projector: Setting wide screen mode")
    
    def off(self):
        print("Projector: Turning off")

class SoundSystem:
    def on(self):
        print("Sound System: Turning on")
    
    def set_surround_sound(self):
        print("Sound System: Setting surround sound")
    
    def set_volume(self, level):
        print(f"Sound System: Setting volume to {level}")
    
    def off(self):
        print("Sound System: Turning off")

class Lights:
    def dim(self, level):
        print(f"Lights: Dimming to {level}%")

# Client has to do ALL this just to watch a movie!
print("=== Watching a movie (WITHOUT Facade) ===")
print("\nStarting movie...")
lights = Lights()
dvd = DVDPlayer()
projector = Projector()
sound = SoundSystem()

lights.dim(10)
projector.on()
projector.wide_screen_mode()
sound.on()
sound.set_surround_sound()
sound.set_volume(5)
dvd.on()
dvd.play("Inception")

print("\nEnding movie...")
dvd.stop()
dvd.off()
sound.off()
projector.off()
lights.dim(100)

print("\nâŒ Too many steps! Client needs to know too much!")
````

## Implementation: Facade Pattern

````python
# Subsystem classes (same as before)
class DVDPlayer:
    def on(self):
        print("  ğŸ“€ DVD Player: ON")
    
    def play(self, movie: str):
        print(f"  â–¶ï¸  DVD Player: Playing '{movie}'")
    
    def stop(self):
        print("  â¹ï¸  DVD Player: Stopped")
    
    def off(self):
        print("  ğŸ“€ DVD Player: OFF")


class Projector:
    def on(self):
        print("  ğŸ“½ï¸  Projector: ON")
    
    def wide_screen_mode(self):
        print("  ğŸ“º Projector: Wide screen mode")
    
    def off(self):
        print("  ğŸ“½ï¸  Projector: OFF")


class SoundSystem:
    def on(self):
        print("  ğŸ”Š Sound System: ON")
    
    def set_surround_sound(self):
        print("  ğŸµ Sound System: Surround sound enabled")
    
    def set_volume(self, level: int):
        print(f"  ğŸ”‰ Sound System: Volume set to {level}")
    
    def off(self):
        print("  ğŸ”Š Sound System: OFF")


class Lights:
    def dim(self, level: int):
        if level < 30:
            print(f"  ğŸŒ™ Lights: Dimmed to {level}% (movie mode)")
        else:
            print(f"  ğŸ’¡ Lights: Brightness at {level}%")


# Facade - Simplifies the interface
class HomeTheaterFacade:
    """Simplified interface for home theater."""
    
    def __init__(self):
        # Initialize all subsystem components
        self.dvd = DVDPlayer()
        self.projector = Projector()
        self.sound = SoundSystem()
        self.lights = Lights()
    
    def watch_movie(self, movie: str) -> None:
        """One method to start watching a movie."""
        print(f"ğŸ¬ Starting movie: {movie}\n")
        
        # Orchestrate all the subsystems
        self.lights.dim(10)
        self.projector.on()
        self.projector.wide_screen_mode()
        self.sound.on()
        self.sound.set_surround_sound()
        self.sound.set_volume(5)
        self.dvd.on()
        self.dvd.play(movie)
        
        print(f"\nâœ… Enjoy your movie!")
    
    def end_movie(self) -> None:
        """One method to end the movie."""
        print("\nğŸ¬ Ending movie...\n")
        
        self.dvd.stop()
        self.dvd.off()
        self.sound.off()
        self.projector.off()
        self.lights.dim(100)
        
        print("\nâœ… Movie ended. Cleanup complete.")


# Demo
print("=== Home Theater with Facade ===")

# Client only needs to know about the facade!
theater = HomeTheaterFacade()

# Simple interface: just two method calls
theater.watch_movie("Inception")
theater.end_movie()

print("\nâœ… Facade simplified the interface!")
````

## Real-World Example: Computer Startup

````python
# Subsystem classes
class CPU:
    def freeze(self) -> None:
        print("  ğŸ§Š CPU: Freezing")
    
    def jump(self, position: int) -> None:
        print(f"  ğŸƒ CPU: Jumping to position {position}")
    
    def execute(self) -> None:
        print("  âš¡ CPU: Executing instructions")


class Memory:
    def load(self, position: int, data: str) -> None:
        print(f"  ğŸ’¾ Memory: Loading '{data}' at position {position}")


class HardDrive:
    def read(self, sector: int, size: int) -> str:
        print(f"  ğŸ’¿ HardDrive: Reading {size} bytes from sector {sector}")
        return "boot_data"


# Facade
class ComputerFacade:
    """Simplified interface for starting a computer."""
    
    def __init__(self):
        self.cpu = CPU()
        self.memory = Memory()
        self.hard_drive = HardDrive()
    
    def start(self) -> None:
        """Start the computer with one method call."""
        print("ğŸ’» Starting computer...\n")
        
        # Complex startup sequence
        self.cpu.freeze()
        boot_data = self.hard_drive.read(0, 1024)
        self.memory.load(0, boot_data)
        self.cpu.jump(0)
        self.cpu.execute()
        
        print("\nâœ… Computer started successfully!")


# Demo
print("\n=== Computer Startup Facade ===")

computer = ComputerFacade()
computer.start()  # One simple method!
````

## Real-World Example: Order Processing System

````python
# Subsystem classes
class InventorySystem:
    def check_stock(self, product_id: str, quantity: int) -> bool:
        print(f"  ğŸ“¦ Inventory: Checking stock for {product_id} (qty: {quantity})")
        return True  # Assume in stock
    
    def reserve_items(self, product_id: str, quantity: int) -> None:
        print(f"  âœ… Inventory: Reserved {quantity} units of {product_id}")


class PaymentSystem:
    def validate_card(self, card_number: str) -> bool:
        print(f"  ğŸ’³ Payment: Validating card {card_number[-4:]}")
        return True
    
    def charge(self, amount: float) -> str:
        print(f"  ğŸ’µ Payment: Charging ${amount:.2f}")
        return "TXN12345"


class ShippingSystem:
    def calculate_shipping(self, weight: float, distance: float) -> float:
        cost = weight * 0.5 + distance * 0.1
        print(f"  ğŸ“® Shipping: Calculated cost ${cost:.2f}")
        return cost
    
    def schedule_delivery(self, address: str) -> str:
        print(f"  ğŸšš Shipping: Scheduled delivery to {address}")
        return "TRACK123"


class NotificationSystem:
    def send_email(self, email: str, message: str) -> None:
        print(f"  ğŸ“§ Notification: Email sent to {email}")
    
    def send_sms(self, phone: str, message: str) -> None:
        print(f"  ğŸ“± Notification: SMS sent to {phone}")


# Facade
class OrderFacade:
    """Simplified interface for placing orders."""
    
    def __init__(self):
        self.inventory = InventorySystem()
        self.payment = PaymentSystem()
        self.shipping = ShippingSystem()
        self.notification = NotificationSystem()
    
    def place_order(
        self,
        product_id: str,
        quantity: int,
        card_number: str,
        amount: float,
        address: str,
        email: str
    ) -> dict:
        """Place an order with one method call."""
        print(f"ğŸ›’ Processing order for {quantity}x {product_id}...\n")
        
        # 1. Check inventory
        if not self.inventory.check_stock(product_id, quantity):
            print("\nâŒ Order failed: Out of stock")
            return {"success": False, "reason": "Out of stock"}
        
        # 2. Reserve items
        self.inventory.reserve_items(product_id, quantity)
        
        # 3. Process payment
        if not self.payment.validate_card(card_number):
            print("\nâŒ Order failed: Invalid card")
            return {"success": False, "reason": "Invalid card"}
        
        transaction_id = self.payment.charge(amount)
        
        # 4. Schedule shipping
        shipping_cost = self.shipping.calculate_shipping(2.5, 100)
        tracking_number = self.shipping.schedule_delivery(address)
        
        # 5. Send notifications
        self.notification.send_email(email, "Order confirmed!")
        
        print("\nâœ… Order placed successfully!")
        return {
            "success": True,
            "transaction_id": transaction_id,
            "tracking_number": tracking_number,
            "total": amount + shipping_cost
        }


# Demo
print("\n=== Order Processing Facade ===")

order_system = OrderFacade()

result = order_system.place_order(
    product_id="LAPTOP123",
    quantity=1,
    card_number="1234-5678-9012-3456",
    amount=999.99,
    address="123 Main St",
    email="customer@example.com"
)

print(f"\nğŸ“Š Order Result:")
print(f"   Transaction ID: {result.get('transaction_id')}")
print(f"   Tracking: {result.get('tracking_number')}")
print(f"   Total: ${result.get('total', 0):.2f}")
````

## Real-World Example: Video Converter

````python
# Complex subsystem
class VideoFile:
    def __init__(self, filename: str):
        self.filename = filename
        self.codec = filename.split(".")[-1]


class CodecFactory:
    @staticmethod
    def extract(file: VideoFile):
        print(f"  ğŸ¥ CodecFactory: Extracting {file.codec} codec")
        return file.codec


class BitrateReader:
    @staticmethod
    def read(filename: str, codec: str):
        print(f"  ğŸ“Š BitrateReader: Reading bitrate from {filename}")
        return "high"


class AudioMixer:
    def fix(self, result):
        print(f"  ğŸµ AudioMixer: Fixing audio")
        return result


# Facade
class VideoConverterFacade:
    """Simplified interface for video conversion."""
    
    def convert(self, filename: str, format: str) -> str:
        """Convert video to target format."""
        print(f"ğŸ¬ Converting {filename} to {format}...\n")
        
        # Complex conversion process
        file = VideoFile(filename)
        codec = CodecFactory.extract(file)
        
        if format == "mp4":
            bitrate = BitrateReader.read(filename, codec)
            result = f"{filename}.mp4"
            mixer = AudioMixer()
            result = mixer.fix(result)
        else:
            result = f"{filename}.{format}"
        
        print(f"\nâœ… Conversion complete: {result}")
        return result


# Demo
print("\n=== Video Converter Facade ===")

converter = VideoConverterFacade()

# Simple interface - client doesn't need to know about codecs, bitrate, etc.
converter.convert("video.avi", "mp4")
print()
converter.convert("movie.mkv", "mp4")
````

## Facade vs Other Patterns

### Facade vs Adapter

**Facade**:
- Simplifies interface
- Works with multiple classes
- Intent: Make subsystem easier to use

**Adapter**:
- Changes interface
- Usually wraps one class
- Intent: Make incompatible interfaces work together

````python
# Facade - Simplifies complex subsystem
class ComplexSubsystemA:
    def operation_a(self):
        return "A"

class ComplexSubsystemB:
    def operation_b(self):
        return "B"

class Facade:
    def __init__(self):
        self.a = ComplexSubsystemA()
        self.b = ComplexSubsystemB()
    
    def simple_operation(self):  # Simplifies multiple operations
        return f"{self.a.operation_a()} + {self.b.operation_b()}"


# Adapter - Changes interface
class OldAPI:
    def old_method(self):
        return "old"

class NewAPI:
    def new_method(self):  # Different interface
        pass

class Adapter(NewAPI):
    def __init__(self, old_api):
        self.old_api = old_api
    
    def new_method(self):  # Adapts old interface to new
        return self.old_api.old_method()


print("Facade: Simplifies complex subsystem")
facade = Facade()
print(f"  Result: {facade.simple_operation()}")

print("\nAdapter: Makes incompatible interfaces work together")
adapter = Adapter(OldAPI())
print(f"  Result: {adapter.new_method()}")
````

## Advantages & Disadvantages

### âœ… Advantages
1. **Simplifies interface**: Easy to use complex subsystem
2. **Decoupling**: Isolates clients from subsystem
3. **Layering**: Create layers in subsystem
4. **Easier testing**: Mock facade instead of entire subsystem
5. **Single point of entry**: Centralized access to subsystem

### âŒ Disadvantages
1. **God object**: Facade can become too large
2. **Not always needed**: Simple subsystems don't need facade
3. **Less flexibility**: Simplified interface may hide useful features

## Common Use Cases

1. **Libraries/frameworks**: Simplify complex APIs
2. **Legacy systems**: Provide modern interface
3. **Microservices**: API gateway as facade
4. **Home automation**: Single control interface
5. **E-commerce**: Order processing
6. **Media systems**: Video/audio conversion
7. **Computer systems**: Startup/shutdown sequences

## Related Patterns

- **Adapter**: Changes interface (Facade simplifies)
- **Mediator**: Centralizes communication (Facade simplifies access)
- **Abstract Factory**: Can be used with Facade to create subsystem objects
- **Singleton**: Facade is often implemented as Singleton

## Best Practices

1. **Keep facade focused**: Don't add business logic
2. **Allow direct access**: Don't force clients to use facade
3. **Multiple facades**: Create specialized facades if needed
4. **Don't hide everything**: Expose advanced features when needed
5. **Document clearly**: Explain what facade simplifies
6. **Avoid God object**: Don't make facade do too much

## Summary

Facade pattern enables:
- Simplified interface to complex subsystem
- Decoupling of clients from implementation
- Easier subsystem usage
- Better code organization

Perfect for: Complex libraries, legacy systems, multi-step processes, home automation.

**Key Insight**: Provide a simple, unified interface that hides the complexity of the subsystem!
