# Abstract Factory Pattern

## Intent
Provide an interface for creating families of related or dependent objects without specifying their concrete classes.

## Problem
You need to create families of related objects that:
- Must be used together (consistency required)
- Have multiple variants/themes
- Should not be mixed between families
- Need to be created without knowing concrete classes

**Real-world analogy**: Furniture store with different styles (Modern, Victorian, ArtDeco) - each style has chair, sofa, table that must match

## When to Use
âœ… **Use when:**
- System needs to be independent of product creation
- Products must be used together (product families)
- Need to enforce family consistency
- Want to provide product library with only interfaces exposed
- Have multiple variants of product families

âŒ **Avoid when:**
- Only single product (use Factory Method instead)
- Products don't need to work together
- Number of product types changes frequently (hard to extend)

## Example 1: Without Abstract Factory

**Problem**: Mixing incompatible product families, no consistency guarantee

````python
# WITHOUT Abstract Factory - Easy to mix incompatible products

class WindowsButton:
    def paint(self):
        print("Rendering Windows button")

class MacButton:
    def paint(self):
        print("Rendering Mac button")

class WindowsCheckbox:
    def paint(self):
        print("Rendering Windows checkbox")

class MacCheckbox:
    def paint(self):
        print("Rendering Mac checkbox")

# Client code - Easy to mix incompatible components!
def create_ui(os_type):
    if os_type == "windows":
        button = WindowsButton()
        checkbox = MacCheckbox()  # Oops! Wrong combination
    else:
        button = MacButton()
        checkbox = WindowsCheckbox()  # Oops! Wrong combination
    
    button.paint()
    checkbox.paint()

create_ui("windows")

print("\nâŒ Easy to mix incompatible components!")
print("âŒ No guarantee of family consistency!")
print("âŒ Client must know all concrete classes!")
````

## Implementation: Abstract Factory Pattern

````python
from abc import ABC, abstractmethod
from typing import List

# Abstract Products
class Button(ABC):
    """Abstract button interface."""
    
    @abstractmethod
    def paint(self) -> None:
        pass
    
    @abstractmethod
    def on_click(self) -> None:
        pass


class Checkbox(ABC):
    """Abstract checkbox interface."""
    
    @abstractmethod
    def paint(self) -> None:
        pass
    
    @abstractmethod
    def on_check(self) -> None:
        pass


# Windows Product Family
class WindowsButton(Button):
    def paint(self) -> None:
        print("  ðŸªŸ Rendering Windows button")
    
    def on_click(self) -> None:
        print("  ðŸ–±ï¸  Windows button clicked")


class WindowsCheckbox(Checkbox):
    def paint(self) -> None:
        print("  ðŸªŸ Rendering Windows checkbox")
    
    def on_check(self) -> None:
        print("  âœ… Windows checkbox checked")


# Mac Product Family
class MacButton(Button):
    def paint(self) -> None:
        print("  ðŸŽ Rendering Mac button")
    
    def on_click(self) -> None:
        print("  ðŸ–±ï¸  Mac button clicked")


class MacCheckbox(Checkbox):
    def paint(self) -> None:
        print("  ðŸŽ Rendering Mac checkbox")
    
    def on_check(self) -> None:
        print("  âœ… Mac checkbox checked")


# Linux Product Family
class LinuxButton(Button):
    def paint(self) -> None:
        print("  ðŸ§ Rendering Linux button")
    
    def on_click(self) -> None:
        print("  ðŸ–±ï¸  Linux button clicked")


class LinuxCheckbox(Checkbox):
    def paint(self) -> None:
        print("  ðŸ§ Rendering Linux checkbox")
    
    def on_check(self) -> None:
        print("  âœ… Linux checkbox checked")


# Abstract Factory
class GUIFactory(ABC):
    """Abstract factory for creating UI component families."""
    
    @abstractmethod
    def create_button(self) -> Button:
        pass
    
    @abstractmethod
    def create_checkbox(self) -> Checkbox:
        pass


# Concrete Factories
class WindowsFactory(GUIFactory):
    """Factory for Windows UI components."""
    
    def create_button(self) -> Button:
        return WindowsButton()
    
    def create_checkbox(self) -> Checkbox:
        return WindowsCheckbox()


class MacFactory(GUIFactory):
    """Factory for Mac UI components."""
    
    def create_button(self) -> Button:
        return MacButton()
    
    def create_checkbox(self) -> Checkbox:
        return MacCheckbox()


class LinuxFactory(GUIFactory):
    """Factory for Linux UI components."""
    
    def create_button(self) -> Button:
        return LinuxButton()
    
    def create_checkbox(self) -> Checkbox:
        return LinuxCheckbox()


# Client Code
class Application:
    """Application that uses abstract factory."""
    
    def __init__(self, factory: GUIFactory):
        self.factory = factory
    
    def create_ui(self) -> None:
        """Create UI using factory - guaranteed consistency."""
        button = self.factory.create_button()
        checkbox = self.factory.create_checkbox()
        
        # All components from same family
        button.paint()
        button.on_click()
        checkbox.paint()
        checkbox.on_check()


# Demo
print("=== Abstract Factory Pattern (Cross-Platform UI) ===")

print("\n1. Windows Application:")
windows_app = Application(WindowsFactory())
windows_app.create_ui()

print("\n2. Mac Application:")
mac_app = Application(MacFactory())
mac_app.create_ui()

print("\n3. Linux Application:")
linux_app = Application(LinuxFactory())
linux_app.create_ui()

print("\nâœ… All UI components guaranteed to be from same family!")
print("âœ… Impossible to mix incompatible components!")
````

## Real-World Example: Furniture Store

````python
# Abstract Products
class Chair(ABC):
    @abstractmethod
    def sit_on(self) -> None:
        pass


class Sofa(ABC):
    @abstractmethod
    def lie_on(self) -> None:
        pass


class CoffeeTable(ABC):
    @abstractmethod
    def place_items(self) -> None:
        pass


# Modern Style Family
class ModernChair(Chair):
    def sit_on(self) -> None:
        print("  ðŸ’º Sitting on sleek modern chair with chrome legs")


class ModernSofa(Sofa):
    def lie_on(self) -> None:
        print("  ðŸ›‹ï¸  Lying on minimalist modern sofa")


class ModernCoffeeTable(CoffeeTable):
    def place_items(self) -> None:
        print("  â˜• Placing items on glass modern coffee table")


# Victorian Style Family
class VictorianChair(Chair):
    def sit_on(self) -> None:
        print("  ðŸ’º Sitting on ornate Victorian chair with carved wood")


class VictorianSofa(Sofa):
    def lie_on(self) -> None:
        print("  ðŸ›‹ï¸  Lying on elegant Victorian sofa with velvet upholstery")


class VictorianCoffeeTable(CoffeeTable):
    def place_items(self) -> None:
        print("  â˜• Placing items on mahogany Victorian coffee table")


# Art Deco Style Family
class ArtDecoChair(Chair):
    def sit_on(self) -> None:
        print("  ðŸ’º Sitting on geometric Art Deco chair")


class ArtDecoSofa(Sofa):
    def lie_on(self) -> None:
        print("  ðŸ›‹ï¸  Lying on luxurious Art Deco sofa")


class ArtDecoCoffeeTable(CoffeeTable):
    def place_items(self) -> None:
        print("  â˜• Placing items on mirrored Art Deco coffee table")


# Abstract Factory
class FurnitureFactory(ABC):
    """Abstract factory for furniture families."""
    
    @abstractmethod
    def create_chair(self) -> Chair:
        pass
    
    @abstractmethod
    def create_sofa(self) -> Sofa:
        pass
    
    @abstractmethod
    def create_coffee_table(self) -> CoffeeTable:
        pass


# Concrete Factories
class ModernFurnitureFactory(FurnitureFactory):
    def create_chair(self) -> Chair:
        return ModernChair()
    
    def create_sofa(self) -> Sofa:
        return ModernSofa()
    
    def create_coffee_table(self) -> CoffeeTable:
        return ModernCoffeeTable()


class VictorianFurnitureFactory(FurnitureFactory):
    def create_chair(self) -> Chair:
        return VictorianChair()
    
    def create_sofa(self) -> Sofa:
        return VictorianSofa()
    
    def create_coffee_table(self) -> CoffeeTable:
        return VictorianCoffeeTable()


class ArtDecoFurnitureFactory(FurnitureFactory):
    def create_chair(self) -> Chair:
        return ArtDecoChair()
    
    def create_sofa(self) -> Sofa:
        return ArtDecoSofa()
    
    def create_coffee_table(self) -> CoffeeTable:
        return ArtDecoCoffeeTable()


# Client Code
def furnish_room(factory: FurnitureFactory) -> None:
    """Furnish a room with matching furniture."""
    chair = factory.create_chair()
    sofa = factory.create_sofa()
    table = factory.create_coffee_table()
    
    # All furniture guaranteed to match
    chair.sit_on()
    sofa.lie_on()
    table.place_items()


# Demo
print("\n=== Furniture Store ===")

print("\n1. Modern Living Room:")
furnish_room(ModernFurnitureFactory())

print("\n2. Victorian Living Room:")
furnish_room(VictorianFurnitureFactory())

print("\n3. Art Deco Living Room:")
furnish_room(ArtDecoFurnitureFactory())

print("\nâœ… All furniture pieces match perfectly!")
````

## Real-World Example: Theme System

````python
# Abstract Products
class Color(ABC):
    @abstractmethod
    def get_rgb(self) -> str:
        pass


class Font(ABC):
    @abstractmethod
    def get_family(self) -> str:
        pass


class Icon(ABC):
    @abstractmethod
    def get_style(self) -> str:
        pass


# Light Theme Family
class LightColor(Color):
    def get_rgb(self) -> str:
        return "rgb(255, 255, 255)"


class LightFont(Font):
    def get_family(self) -> str:
        return "Helvetica, sans-serif"


class LightIcon(Icon):
    def get_style(self) -> str:
        return "outlined"


# Dark Theme Family
class DarkColor(Color):
    def get_rgb(self) -> str:
        return "rgb(30, 30, 30)"


class DarkFont(Font):
    def get_family(self) -> str:
        return "Roboto Mono, monospace"


class DarkIcon(Icon):
    def get_style(self) -> str:
        return "filled"


# High Contrast Theme Family
class HighContrastColor(Color):
    def get_rgb(self) -> str:
        return "rgb(0, 0, 0)"


class HighContrastFont(Font):
    def get_family(self) -> str:
        return "Arial Black, sans-serif"


class HighContrastIcon(Icon):
    def get_style(self) -> str:
        return "bold"


# Abstract Factory
class ThemeFactory(ABC):
    """Abstract factory for theme components."""
    
    @abstractmethod
    def create_color(self) -> Color:
        pass
    
    @abstractmethod
    def create_font(self) -> Font:
        pass
    
    @abstractmethod
    def create_icon(self) -> Icon:
        pass


# Concrete Factories
class LightThemeFactory(ThemeFactory):
    def create_color(self) -> Color:
        return LightColor()
    
    def create_font(self) -> Font:
        return LightFont()
    
    def create_icon(self) -> Icon:
        return LightIcon()


class DarkThemeFactory(ThemeFactory):
    def create_color(self) -> Color:
        return DarkColor()
    
    def create_font(self) -> Font:
        return DarkFont()
    
    def create_icon(self) -> Icon:
        return DarkIcon()


class HighContrastThemeFactory(ThemeFactory):
    def create_color(self) -> Color:
        return HighContrastColor()
    
    def create_font(self) -> Font:
        return HighContrastFont()
    
    def create_icon(self) -> Icon:
        return HighContrastIcon()


# Client Code
class WebPage:
    """Web page that applies theme."""
    
    def __init__(self, theme_factory: ThemeFactory):
        self.color = theme_factory.create_color()
        self.font = theme_factory.create_font()
        self.icon = theme_factory.create_icon()
    
    def render(self) -> None:
        """Render page with theme."""
        print(f"  ðŸŽ¨ Background: {self.color.get_rgb()}")
        print(f"  ðŸ”¤ Font: {self.font.get_family()}")
        print(f"  ðŸŽ¯ Icons: {self.icon.get_style()}")


# Demo
print("\n=== Theme System ===")

print("\n1. Light Theme:")
light_page = WebPage(LightThemeFactory())
light_page.render()

print("\n2. Dark Theme:")
dark_page = WebPage(DarkThemeFactory())
dark_page.render()

print("\n3. High Contrast Theme:")
hc_page = WebPage(HighContrastThemeFactory())
hc_page.render()

print("\nâœ… All theme components work together harmoniously!")
````

## Real-World Example: Document Exporters

````python
# Abstract Products
class DocumentFormatter(ABC):
    @abstractmethod
    def format_header(self, text: str) -> str:
        pass
    
    @abstractmethod
    def format_paragraph(self, text: str) -> str:
        pass


class ImageRenderer(ABC):
    @abstractmethod
    def render_image(self, url: str) -> str:
        pass


class TableRenderer(ABC):
    @abstractmethod
    def render_table(self, rows: int, cols: int) -> str:
        pass


# HTML Export Family
class HTMLFormatter(DocumentFormatter):
    def format_header(self, text: str) -> str:
        return f"<h1>{text}</h1>"
    
    def format_paragraph(self, text: str) -> str:
        return f"<p>{text}</p>"


class HTMLImageRenderer(ImageRenderer):
    def render_image(self, url: str) -> str:
        return f'<img src="{url}" />'


class HTMLTableRenderer(TableRenderer):
    def render_table(self, rows: int, cols: int) -> str:
        return f"<table><tr><td>Table {rows}x{cols}</td></tr></table>"


# Markdown Export Family
class MarkdownFormatter(DocumentFormatter):
    def format_header(self, text: str) -> str:
        return f"# {text}"
    
    def format_paragraph(self, text: str) -> str:
        return f"\n{text}\n"


class MarkdownImageRenderer(ImageRenderer):
    def render_image(self, url: str) -> str:
        return f"![Image]({url})"


class MarkdownTableRenderer(TableRenderer):
    def render_table(self, rows: int, cols: int) -> str:
        return f"| Table {rows}x{cols} |\n|---|\n| Data |"


# PDF Export Family
class PDFFormatter(DocumentFormatter):
    def format_header(self, text: str) -> str:
        return f"[PDF Header: {text}]"
    
    def format_paragraph(self, text: str) -> str:
        return f"[PDF Paragraph: {text}]"


class PDFImageRenderer(ImageRenderer):
    def render_image(self, url: str) -> str:
        return f"[PDF Image: {url}]"


class PDFTableRenderer(TableRenderer):
    def render_table(self, rows: int, cols: int) -> str:
        return f"[PDF Table: {rows}x{cols}]"


# Abstract Factory
class DocumentExportFactory(ABC):
    """Abstract factory for document export components."""
    
    @abstractmethod
    def create_formatter(self) -> DocumentFormatter:
        pass
    
    @abstractmethod
    def create_image_renderer(self) -> ImageRenderer:
        pass
    
    @abstractmethod
    def create_table_renderer(self) -> TableRenderer:
        pass


# Concrete Factories
class HTMLExportFactory(DocumentExportFactory):
    def create_formatter(self) -> DocumentFormatter:
        return HTMLFormatter()
    
    def create_image_renderer(self) -> ImageRenderer:
        return HTMLImageRenderer()
    
    def create_table_renderer(self) -> TableRenderer:
        return HTMLTableRenderer()


class MarkdownExportFactory(DocumentExportFactory):
    def create_formatter(self) -> DocumentFormatter:
        return MarkdownFormatter()
    
    def create_image_renderer(self) -> ImageRenderer:
        return MarkdownImageRenderer()
    
    def create_table_renderer(self) -> TableRenderer:
        return MarkdownTableRenderer()


class PDFExportFactory(DocumentExportFactory):
    def create_formatter(self) -> DocumentFormatter:
        return PDFFormatter()
    
    def create_image_renderer(self) -> ImageRenderer:
        return PDFImageRenderer()
    
    def create_table_renderer(self) -> TableRenderer:
        return PDFTableRenderer()


# Client Code
class DocumentExporter:
    """Exports documents using factory."""
    
    def __init__(self, factory: DocumentExportFactory):
        self.formatter = factory.create_formatter()
        self.image_renderer = factory.create_image_renderer()
        self.table_renderer = factory.create_table_renderer()
    
    def export(self) -> None:
        """Export document with all components."""
        print(f"  {self.formatter.format_header('My Document')}")
        print(f"  {self.formatter.format_paragraph('This is content.')}")
        print(f"  {self.image_renderer.render_image('photo.jpg')}")
        print(f"  {self.table_renderer.render_table(3, 2)}")


# Demo
print("\n=== Document Export System ===")

print("\n1. Export as HTML:")
html_exporter = DocumentExporter(HTMLExportFactory())
html_exporter.export()

print("\n2. Export as Markdown:")
md_exporter = DocumentExporter(MarkdownExportFactory())
md_exporter.export()

print("\n3. Export as PDF:")
pdf_exporter = DocumentExporter(PDFExportFactory())
pdf_exporter.export()

print("\nâœ… All export components guaranteed to be compatible!")
````

## Advantages & Disadvantages

### âœ… Advantages
1. **Product family consistency**: Ensures products work together
2. **Isolation of concrete classes**: Clients work with interfaces
3. **Easy to swap families**: Change entire product family at once
4. **Single Responsibility**: Separates product creation code
5. **Open/Closed Principle**: Easy to add new product families

### âŒ Disadvantages
1. **Complexity**: Many classes and interfaces
2. **Hard to extend products**: Adding new product type requires updating all factories
3. **Overhead**: May be overkill for simple cases
4. **Rigid structure**: Product families must be known upfront

## Abstract Factory vs Factory Method

**Abstract Factory**:
- Creates families of related products
- Multiple factory methods (one per product)
- Uses composition
- Focus on product compatibility

**Factory Method**:
- Creates single product
- One factory method
- Uses inheritance
- Focus on object creation

**Example**:
```python
# Factory Method - creates one thing
class ButtonFactory:
    def create_button() -> Button

# Abstract Factory - creates family of things
class UIFactory:
    def create_button() -> Button
    def create_checkbox() -> Checkbox
    def create_textbox() -> TextBox
```

## Common Use Cases

1. **Cross-platform UI**: Different OS widget families
2. **Themes**: Light/Dark/Custom theme components
3. **Database drivers**: Different DB client families (connection, command, reader)
4. **Document formats**: HTML/PDF/Markdown exporters
5. **Game environments**: Medieval/SciFi/Fantasy asset sets
6. **Payment gateways**: Different payment processor families
7. **Cloud providers**: AWS/Azure/GCP service families

## Related Patterns

- **Factory Method**: Abstract Factory often implemented with Factory Methods
- **Singleton**: Factories often implemented as Singletons
- **Prototype**: Alternative to Abstract Factory (uses cloning)
- **Facade**: Can provide simple interface to complex factory

## Best Practices

1. **Use for product families**: Only when products must work together
2. **Keep products focused**: Don't bloat with too many products
3. **Default factory**: Provide sensible default if possible
4. **Factory selection**: Use factory method or config to select factory
5. **Product interfaces**: Keep product interfaces stable
6. **Naming convention**: `create_X()` methods clarify intent
7. **Consider Builder**: If products need complex configuration

## Python-Specific: Dynamic Factory Selection

Python allows dynamic factory selection:

```python
# Registry pattern with Abstract Factory
FACTORIES = {
    'windows': WindowsFactory,
    'mac': MacFactory,
    'linux': LinuxFactory
}

def get_factory(platform: str) -> GUIFactory:
    return FACTORIES[platform]()

# Use it
factory = get_factory('mac')
app = Application(factory)
```

## Summary

Abstract Factory pattern enables:
- Creating families of related objects
- Guaranteeing product compatibility
- Swapping entire product families
- Isolating concrete classes from clients

Perfect for: Cross-platform applications, themes, multi-format exporters, database drivers, game environments.

**Key Insight**: Create families of related objects through a factory interface, ensuring products work together and making it easy to swap entire product families!
