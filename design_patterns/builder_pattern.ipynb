# Builder Pattern

## Intent
Separate the construction of a complex object from its representation, allowing the same construction process to create different representations.

## Problem
You need to construct complex objects that:
- Have many optional parameters (telescoping constructor problem)
- Require step-by-step construction
- Can have different representations
- Need validation during construction

**Real-world analogy**: Building a house - same construction process (foundation, walls, roof) creates different house types (wooden, stone, modern)

## When to Use
âœ… **Use when:**
- Object construction is complex with many steps
- Many optional parameters (avoid telescoping constructors)
- Different representations of same object needed
- Want to enforce construction order
- Need immutable objects with many fields

âŒ **Avoid when:**
- Object is simple with few parameters
- Construction has no meaningful steps
- Only one representation exists

## Example 1: Without Builder

**Problem**: Telescoping constructor, hard to read, easy to make mistakes

````python
# WITHOUT Builder - Telescoping constructor problem

class Pizza:
    def __init__(self, size, cheese=False, pepperoni=False, 
                 mushrooms=False, olives=False, bacon=False, 
                 onions=False, extra_sauce=False):
        self.size = size
        self.cheese = cheese
        self.pepperoni = pepperoni
        self.mushrooms = mushrooms
        self.olives = olives
        self.bacon = bacon
        self.onions = onions
        self.extra_sauce = extra_sauce

# Creating pizza is confusing and error-prone
pizza1 = Pizza("large", True, True, False, True, False, False, True)
# Which parameter is which? Hard to tell!

# Using named parameters helps but still verbose
pizza2 = Pizza(
    size="medium",
    cheese=True,
    pepperoni=True,
    mushrooms=False,
    olives=True,
    bacon=False,
    onions=False,
    extra_sauce=True
)

print("\nâŒ Too many constructor parameters!")
print("âŒ Easy to make mistakes with parameter order!")
print("âŒ Hard to read and understand!")
````

## Implementation: Builder Pattern

````python
from abc import ABC, abstractmethod
from typing import Optional

# Product
class House:
    """Complex product being built."""
    
    def __init__(self):
        self.foundation = None
        self.structure = None
        self.roof = None
        self.interior = None
        self.windows = 0
        self.garage = False
        self.pool = False
    
    def __str__(self) -> str:
        parts = [
            f"Foundation: {self.foundation}",
            f"Structure: {self.structure}",
            f"Roof: {self.roof}",
            f"Interior: {self.interior}",
            f"Windows: {self.windows}",
            f"Garage: {self.garage}",
            f"Pool: {self.pool}"
        ]
        return "\n    ".join(parts)


# Builder Interface
class HouseBuilder(ABC):
    """Abstract builder interface."""
    
    def __init__(self):
        self.house = House()
    
    @abstractmethod
    def build_foundation(self) -> None:
        pass
    
    @abstractmethod
    def build_structure(self) -> None:
        pass
    
    @abstractmethod
    def build_roof(self) -> None:
        pass
    
    @abstractmethod
    def build_interior(self) -> None:
        pass
    
    def add_windows(self, count: int) -> None:
        self.house.windows = count
    
    def add_garage(self) -> None:
        self.house.garage = True
    
    def add_pool(self) -> None:
        self.house.pool = True
    
    def get_house(self) -> House:
        return self.house


# Concrete Builder 1
class WoodenHouseBuilder(HouseBuilder):
    """Builder for wooden houses."""
    
    def build_foundation(self) -> None:
        print("  ðŸ”¨ Building wooden foundation")
        self.house.foundation = "Wooden piles"
    
    def build_structure(self) -> None:
        print("  ðŸªµ Building wooden structure")
        self.house.structure = "Wooden frame"
    
    def build_roof(self) -> None:
        print("  ðŸ  Building wooden roof")
        self.house.roof = "Wooden shingles"
    
    def build_interior(self) -> None:
        print("  ðŸŽ¨ Finishing wooden interior")
        self.house.interior = "Wooden panels"


# Concrete Builder 2
class StoneHouseBuilder(HouseBuilder):
    """Builder for stone houses."""
    
    def build_foundation(self) -> None:
        print("  ðŸ”¨ Building stone foundation")
        self.house.foundation = "Concrete foundation"
    
    def build_structure(self) -> None:
        print("  ðŸ§± Building stone structure")
        self.house.structure = "Stone walls"
    
    def build_roof(self) -> None:
        print("  ðŸ  Building stone roof")
        self.house.roof = "Stone tiles"
    
    def build_interior(self) -> None:
        print("  ðŸŽ¨ Finishing stone interior")
        self.house.interior = "Stone accents"


# Concrete Builder 3
class ModernHouseBuilder(HouseBuilder):
    """Builder for modern houses."""
    
    def build_foundation(self) -> None:
        print("  ðŸ”¨ Building modern foundation")
        self.house.foundation = "Reinforced concrete"
    
    def build_structure(self) -> None:
        print("  ðŸ—ï¸  Building modern structure")
        self.house.structure = "Steel frame with glass"
    
    def build_roof(self) -> None:
        print("  ðŸ  Building modern roof")
        self.house.roof = "Flat roof with solar panels"
    
    def build_interior(self) -> None:
        print("  ðŸŽ¨ Finishing modern interior")
        self.house.interior = "Minimalist design"


# Director
class ConstructionDirector:
    """Director controls building process."""
    
    def __init__(self, builder: HouseBuilder):
        self.builder = builder
    
    def construct_basic_house(self) -> House:
        """Build basic house."""
        print("  ðŸ“‹ Building basic house...")
        self.builder.build_foundation()
        self.builder.build_structure()
        self.builder.build_roof()
        self.builder.add_windows(4)
        return self.builder.get_house()
    
    def construct_full_house(self) -> House:
        """Build full-featured house."""
        print("  ðŸ“‹ Building full-featured house...")
        self.builder.build_foundation()
        self.builder.build_structure()
        self.builder.build_roof()
        self.builder.build_interior()
        self.builder.add_windows(10)
        self.builder.add_garage()
        self.builder.add_pool()
        return self.builder.get_house()


# Demo
print("=== Builder Pattern (House Construction) ===")

print("\n1. Basic Wooden House:")
wooden_builder = WoodenHouseBuilder()
director = ConstructionDirector(wooden_builder)
basic_wooden = director.construct_basic_house()
print(f"\n  ðŸ¡ Result:\n    {basic_wooden}")

print("\n2. Full Stone House:")
stone_builder = StoneHouseBuilder()
director = ConstructionDirector(stone_builder)
full_stone = director.construct_full_house()
print(f"\n  ðŸ° Result:\n    {full_stone}")

print("\n3. Custom Modern House (no director):")
modern_builder = ModernHouseBuilder()
modern_builder.build_foundation()
modern_builder.build_structure()
modern_builder.build_roof()
modern_builder.build_interior()
modern_builder.add_windows(20)
modern_builder.add_garage()
# No pool
custom_modern = modern_builder.get_house()
print(f"\n  ðŸ¢ Result:\n    {custom_modern}")

print("\nâœ… Same construction process, different representations!")
print("âœ… Can use director for common configurations!")
print("âœ… Can build custom configurations without director!")
````

## Real-World Example: Meal Builder

````python
# Product
class Meal:
    """Complex meal being built."""
    
    def __init__(self):
        self.burger = None
        self.drink = None
        self.side = None
        self.dessert = None
    
    def get_price(self) -> float:
        price = 0.0
        if self.burger: price += self.burger[1]
        if self.drink: price += self.drink[1]
        if self.side: price += self.side[1]
        if self.dessert: price += self.dessert[1]
        return price
    
    def __str__(self) -> str:
        items = []
        if self.burger: items.append(f"Burger: {self.burger[0]} (${self.burger[1]})")
        if self.drink: items.append(f"Drink: {self.drink[0]} (${self.drink[1]})")
        if self.side: items.append(f"Side: {self.side[0]} (${self.side[1]})")
        if self.dessert: items.append(f"Dessert: {self.dessert[0]} (${self.dessert[1]})")
        total = f"Total: ${self.get_price():.2f}"
        return "\n    ".join(items + [total])


# Builder Interface
class MealBuilder(ABC):
    """Abstract meal builder."""
    
    def __init__(self):
        self.meal = Meal()
    
    @abstractmethod
    def add_burger(self) -> 'MealBuilder':
        pass
    
    @abstractmethod
    def add_drink(self) -> 'MealBuilder':
        pass
    
    @abstractmethod
    def add_side(self) -> 'MealBuilder':
        pass
    
    def add_dessert(self) -> 'MealBuilder':
        # Optional for all meal types
        return self
    
    def build(self) -> Meal:
        return self.meal


# Concrete Builder 1
class VeggieMealBuilder(MealBuilder):
    """Builder for vegetarian meals."""
    
    def add_burger(self) -> 'VeggieMealBuilder':
        self.meal.burger = ("Veggie Burger", 5.99)
        return self
    
    def add_drink(self) -> 'VeggieMealBuilder':
        self.meal.drink = ("Fresh Juice", 2.99)
        return self
    
    def add_side(self) -> 'VeggieMealBuilder':
        self.meal.side = ("Sweet Potato Fries", 3.49)
        return self
    
    def add_dessert(self) -> 'VeggieMealBuilder':
        self.meal.dessert = ("Fruit Salad", 3.99)
        return self


# Concrete Builder 2
class ChickenMealBuilder(MealBuilder):
    """Builder for chicken meals."""
    
    def add_burger(self) -> 'ChickenMealBuilder':
        self.meal.burger = ("Chicken Burger", 6.99)
        return self
    
    def add_drink(self) -> 'ChickenMealBuilder':
        self.meal.drink = ("Soda", 1.99)
        return self
    
    def add_side(self) -> 'ChickenMealBuilder':
        self.meal.side = ("Regular Fries", 2.49)
        return self
    
    def add_dessert(self) -> 'ChickenMealBuilder':
        self.meal.dessert = ("Ice Cream", 2.99)
        return self


# Concrete Builder 3
class BeefMealBuilder(MealBuilder):
    """Builder for beef meals."""
    
    def add_burger(self) -> 'BeefMealBuilder':
        self.meal.burger = ("Beef Burger", 7.99)
        return self
    
    def add_drink(self) -> 'BeefMealBuilder':
        self.meal.drink = ("Milkshake", 3.99)
        return self
    
    def add_side(self) -> 'BeefMealBuilder':
        self.meal.side = ("Onion Rings", 3.99)
        return self
    
    def add_dessert(self) -> 'BeefMealBuilder':
        self.meal.dessert = ("Apple Pie", 3.49)
        return self


# Demo
print("\n=== Meal Builder ===")

print("\n1. Veggie Combo Meal:")
veggie_meal = VeggieMealBuilder() \
    .add_burger() \
    .add_drink() \
    .add_side() \
    .build()
print(f"  ðŸ¥— {veggie_meal}")

print("\n2. Chicken Meal with Dessert:")
chicken_meal = ChickenMealBuilder() \
    .add_burger() \
    .add_drink() \
    .add_side() \
    .add_dessert() \
    .build()
print(f"  ðŸ— {chicken_meal}")

print("\n3. Custom Beef Meal (burger + drink only):")
custom_meal = BeefMealBuilder() \
    .add_burger() \
    .add_drink() \
    .build()
print(f"  ðŸ” {custom_meal}")

print("\nâœ… Fluent interface makes building easy!")
print("âœ… Can customize meal composition!")
````

## Real-World Example: SQL Query Builder

````python
from typing import List, Dict, Any

# Product
class SQLQuery:
    """SQL query being built."""
    
    def __init__(self):
        self.select_fields: List[str] = []
        self.from_table: str = None
        self.where_conditions: List[str] = []
        self.order_by_fields: List[str] = []
        self.limit_value: int = None
        self.joins: List[str] = []
    
    def to_sql(self) -> str:
        """Generate SQL string."""
        parts = []
        
        # SELECT
        if self.select_fields:
            parts.append(f"SELECT {', '.join(self.select_fields)}")
        else:
            parts.append("SELECT *")
        
        # FROM
        if self.from_table:
            parts.append(f"FROM {self.from_table}")
        
        # JOIN
        for join in self.joins:
            parts.append(join)
        
        # WHERE
        if self.where_conditions:
            parts.append(f"WHERE {' AND '.join(self.where_conditions)}")
        
        # ORDER BY
        if self.order_by_fields:
            parts.append(f"ORDER BY {', '.join(self.order_by_fields)}")
        
        # LIMIT
        if self.limit_value:
            parts.append(f"LIMIT {self.limit_value}")
        
        return "\n".join(parts)


# Builder
class QueryBuilder:
    """Fluent SQL query builder."""
    
    def __init__(self):
        self.query = SQLQuery()
    
    def select(self, *fields: str) -> 'QueryBuilder':
        """Add SELECT fields."""
        self.query.select_fields.extend(fields)
        return self
    
    def from_table(self, table: str) -> 'QueryBuilder':
        """Set FROM table."""
        self.query.from_table = table
        return self
    
    def where(self, condition: str) -> 'QueryBuilder':
        """Add WHERE condition."""
        self.query.where_conditions.append(condition)
        return self
    
    def join(self, table: str, on: str) -> 'QueryBuilder':
        """Add JOIN clause."""
        self.query.joins.append(f"JOIN {table} ON {on}")
        return self
    
    def order_by(self, *fields: str) -> 'QueryBuilder':
        """Add ORDER BY fields."""
        self.query.order_by_fields.extend(fields)
        return self
    
    def limit(self, value: int) -> 'QueryBuilder':
        """Set LIMIT."""
        self.query.limit_value = value
        return self
    
    def build(self) -> SQLQuery:
        """Get final query."""
        return self.query


# Demo
print("\n=== SQL Query Builder ===")

print("\n1. Simple SELECT query:")
query1 = QueryBuilder() \
    .select("name", "email") \
    .from_table("users") \
    .where("age > 18") \
    .limit(10) \
    .build()
print(f"\n{query1.to_sql()}\n")

print("\n2. Complex JOIN query:")
query2 = QueryBuilder() \
    .select("u.name", "u.email", "o.order_date", "o.total") \
    .from_table("users u") \
    .join("orders o", "u.id = o.user_id") \
    .where("o.total > 100") \
    .where("o.status = 'completed'") \
    .order_by("o.order_date DESC") \
    .limit(20) \
    .build()
print(f"\n{query2.to_sql()}\n")

print("\n3. Simple query without WHERE:")
query3 = QueryBuilder() \
    .from_table("products") \
    .order_by("price DESC") \
    .limit(5) \
    .build()
print(f"\n{query3.to_sql()}\n")

print("âœ… Clean, readable query construction!")
print("âœ… Fluent interface is self-documenting!")
````

## Real-World Example: Report Builder

````python
from datetime import datetime

# Product
class Report:
    """Report being built."""
    
    def __init__(self, format_type: str):
        self.format_type = format_type
        self.title = None
        self.header = None
        self.content = []
        self.footer = None
        self.metadata = {}
    
    def export(self) -> str:
        """Export report in specified format."""
        if self.format_type == "html":
            return self._export_html()
        elif self.format_type == "markdown":
            return self._export_markdown()
        else:
            return self._export_text()
    
    def _export_html(self) -> str:
        lines = ["<html><body>"]
        if self.title:
            lines.append(f"<h1>{self.title}</h1>")
        if self.header:
            lines.append(f"<p><i>{self.header}</i></p>")
        for item in self.content:
            lines.append(f"<p>{item}</p>")
        if self.footer:
            lines.append(f"<footer>{self.footer}</footer>")
        lines.append("</body></html>")
        return "\n".join(lines)
    
    def _export_markdown(self) -> str:
        lines = []
        if self.title:
            lines.append(f"# {self.title}\n")
        if self.header:
            lines.append(f"*{self.header}*\n")
        for item in self.content:
            lines.append(f"{item}\n")
        if self.footer:
            lines.append(f"\n---\n{self.footer}")
        return "\n".join(lines)
    
    def _export_text(self) -> str:
        lines = []
        if self.title:
            lines.append(self.title)
            lines.append("=" * len(self.title))
        if self.header:
            lines.append(self.header)
        lines.extend(self.content)
        if self.footer:
            lines.append("")
            lines.append(self.footer)
        return "\n".join(lines)


# Builder
class ReportBuilder:
    """Fluent report builder."""
    
    def __init__(self, format_type: str = "text"):
        self.report = Report(format_type)
    
    def set_title(self, title: str) -> 'ReportBuilder':
        self.report.title = title
        return self
    
    def set_header(self, header: str) -> 'ReportBuilder':
        self.report.header = header
        return self
    
    def add_section(self, content: str) -> 'ReportBuilder':
        self.report.content.append(content)
        return self
    
    def set_footer(self, footer: str) -> 'ReportBuilder':
        self.report.footer = footer
        return self
    
    def add_metadata(self, key: str, value: Any) -> 'ReportBuilder':
        self.report.metadata[key] = value
        return self
    
    def build(self) -> Report:
        return self.report


# Demo
print("\n=== Report Builder ===")

# Build same content in different formats
title = "Monthly Sales Report"
header = f"Generated on {datetime.now().strftime('%Y-%m-%d')}"
sections = [
    "Total sales: $125,000",
    "New customers: 45",
    "Growth: +15%"
]
footer = "Confidential - Internal Use Only"

print("\n1. HTML Report:")
html_builder = ReportBuilder("html")
html_report = html_builder \
    .set_title(title) \
    .set_header(header) \
    .add_section(sections[0]) \
    .add_section(sections[1]) \
    .add_section(sections[2]) \
    .set_footer(footer) \
    .build()
print(html_report.export())

print("\n2. Markdown Report:")
md_builder = ReportBuilder("markdown")
md_report = md_builder \
    .set_title(title) \
    .set_header(header) \
    .add_section(sections[0]) \
    .add_section(sections[1]) \
    .add_section(sections[2]) \
    .set_footer(footer) \
    .build()
print(md_report.export())

print("\n3. Plain Text Report:")
text_builder = ReportBuilder("text")
text_report = text_builder \
    .set_title(title) \
    .set_header(header) \
    .add_section(sections[0]) \
    .add_section(sections[1]) \
    .add_section(sections[2]) \
    .set_footer(footer) \
    .build()
print(text_report.export())

print("\nâœ… Same content, different formats!")
````

## Advantages & Disadvantages

### âœ… Advantages
1. **Avoid telescoping constructors**: No complex parameter lists
2. **Step-by-step construction**: Build objects incrementally
3. **Different representations**: Same process, different products
4. **Immutability**: Easy to create immutable objects
5. **Readable code**: Fluent interface is self-documenting
6. **Validation**: Can validate during construction
7. **Reusability**: Directors encapsulate common configurations

### âŒ Disadvantages
1. **Complexity**: More classes and code
2. **Overhead**: May be overkill for simple objects
3. **Code duplication**: Similar builders may duplicate code
4. **Learning curve**: Developers must understand pattern

## Builder vs Factory Patterns

**Builder**:
- Constructs complex objects step-by-step
- Focus on HOW object is created (process)
- Returns object after multiple steps
- Good for objects with many optional parameters

**Factory Method/Abstract Factory**:
- Creates objects in one step
- Focus on WHAT object is created (type)
- Returns object immediately
- Good for creating families or variants

**Example**:
```python
# Factory - creates in one step
house = HouseFactory.create("modern")

# Builder - creates step by step
house = HouseBuilder() \
    .set_foundation("concrete") \
    .set_walls("brick") \
    .set_roof("tile") \
    .build()
```

## Common Use Cases

1. **Complex objects**: Houses, meals, reports with many parts
2. **Query builders**: SQL, NoSQL, API query construction
3. **Document builders**: HTML, XML, JSON generation
4. **Test data builders**: Creating test objects with defaults
5. **Configuration**: Complex configuration objects
6. **UI builders**: Form builders, dialog builders
7. **Immutable objects**: Objects with many fields that must be immutable

## Related Patterns

- **Abstract Factory**: Can use Builder internally
- **Composite**: Builders often build Composite structures
- **Fluent Interface**: Often used with Builder for readability
- **Prototype**: Alternative to Builder for object creation

## Best Practices

1. **Fluent interface**: Return `self` for method chaining
2. **Required vs optional**: Make required fields explicit
3. **Validation**: Validate in `build()` method
4. **Immutability**: Make product immutable after building
5. **Director optional**: Can use builder directly for flexibility
6. **Reset method**: Reuse builder for multiple objects
7. **Sensible defaults**: Provide defaults for optional fields
8. **Build once**: Prevent calling `build()` multiple times

## Python-Specific: Using dataclasses

Python's `dataclasses` can reduce boilerplate:


## Summary

Builder pattern enables:
- Step-by-step construction of complex objects
- Avoiding telescoping constructors
- Creating different representations with same process
- Building immutable objects with many fields

Perfect for: Complex objects, query builders, document generators, test data, configuration, immutable objects.

**Key Insight**: Separate complex object construction from representation, using step-by-step building process for clarity and flexibility!
