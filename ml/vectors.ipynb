Basic Arithmetic Operations on Vectors

```python
list1 = [1, 2, 3]
list2 = [4, 5, 6]

vector1 = np.array(list1)

print("First Vector:" + str(vector1))

vector2 = np.array(list2)

print("Second Vector:" + str(vector2))

addition = vector1 + vector2

print("Addition of Vectors:" + str(addition))

subtraction = vector1 - vector2

print("Subtraction of Vectors:" + str(subtraction))

multiplication = vector1 * vector2

print("Multiplication of Vectors:" + str(multiplication))

division = vector1 / vector2

print("Division of Vectors:" + str(division))
```

```python
import numpy as np

list1 = [5, 6, 9]
list2 = [1, 2, 3]

vector1 = np.array(list1)
print("First Vector  : " + str(vector1))

vector2 = np.array(list2)
print("Second Vector : " + str(vector2))

dot_product = np.dot(vector1, vector2)
print("Dot Product of Vectors: " + str(dot_product))
```

```python
import numpy as np

list1 = [1, 2, 3]

vector = np.array(list1)
print("Vector  : " + str(vector))

scalar = 3
print("Scalar : " + str(scalar))

scalar_multiplication = vector * scalar
print("Scalar Multiplication : " + str(scalar_multiplication))
```
